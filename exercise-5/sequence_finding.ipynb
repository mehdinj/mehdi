```python
# This sccript searches for uniqe contig names and number of sequence AATG in their respective sequences and outputs it
# as a csv file 

# This script goes through a loop of every line of the data file using file object of the data file
# If the line selected has the contig name in it, it will first outputs the last contig and AATG value and then resets
# the contig name.
# If the selected line does not contain the contig name, it will search for AATG sequence and adds the number of AATG 
# sequences found to the variable value to store and count the number of AATG sequences in the uniqe contig segment.

file_object=open("/home/eeb177-student/Desktop/eeb-177/CSB/unix/data/Marra2014_data.fasta")
output=open("output.csv","w")

import re
string=""
my_dict={}
value=0
contig_name="contig00001"

for line in file_object:
    if (re.match(pattern=">", string=line)):
        
        output.write(contig_name+","+str(value)+"\n")
        
        contig_name = re.search(">(\w*)\s.*", line).group(1)
        
        value=0
    
    else:
            value=value+ len(re.findall(pattern="AATG|aatg", string=line))
    
          
file_object.close()
output.close()
    
#part of the output file (output.csv); values are equal to the solution file
# contig00001,0
# contig00001,4
# contig00002,1
# contig00003,1
# contig00004,1
# contig00005,4
# contig00006,9
#  ...
```
