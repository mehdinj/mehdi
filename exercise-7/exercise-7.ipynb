```python
# load in the file

in_file=open("cocoli.txt")
#read in the first line 
#its a header that we don't want in dictionaries 
in_file.readline()
# set up empty dictionaries
size_94={}
size_97={}
size_98={}

for line in in_file:
    
    items=line.split()#default is tab
    tree_id=items[0]
    size_in_94= float(items[4])
    size_in_97= float(items[5])
    size_in_98= float(items[6])
    
    size_94[tree_id]= size_in_94
    size_97[tree_id]= size_in_97
    size_98[tree_id]= size_in_98

```

```python
import matplotlib.pyplot as plt

# plot some histograms

sizes_in_94=list(size_94.values())

plt.hist(sizes_in_94 , log=True)
plt.show()
```

```python
f, (ax1,ax2)= plt.subplots(1,2, sharex= True)
ax1.hist(sizes_in_94, log= False)
ax1.set_title('Unlogged')
ax2.hist(sizes_in_94, log=True)
ax2.set_title('Logged')

plt.show()

```

```python
temp=[]
temp[:]= (value for value in sizes_in_94 if value != -1 and value != -2)

f, (ax1,ax2)= plt.subplots(1,2, sharex= True)
ax1.hist(temp, log= False)
ax1.set_title('Unlogged - removed -1 and -2')
ax2.hist(temp, log=True)
ax2.set_title('Logged - removed -1 and -2')

plt.show()
```
